```python
#Exercicio 2

def celsiusToFahrenheit(temp):
    return temp * (9.0 / 5.0) + 32.0

temp = float(input("Digite a temperatura em Celsius: "))

print(f"A temperatura em Fahrenheit é {celsiusToFahrenheit(temp):.2f}")
```

```python
#Exercicio 3

def dollarToReal(value, exchangeRate):
    return value * exchangeRate

value = float(input("Digite o valor em Dolar: "))
exchangeRate = float(input("Digite a cotação do Dolar: "))

print(f"R${dollarToReal(value, exchangeRate):.2f}")
```

```python
#Exercicio 4

def paymentCalc(days, payment, discount):
    return days * payment * (1 - (discount / 100))

days = float(input("Digite o numero de dias trabalhados: "))

print(f"Pagamento: R${paymentCalc(days, payment=30, discount=8):.2f}")
```

```python
#Exercicio 5

import numpy as np

def triangleAreaCalc(values):
    s = (sum(values)) / 2
    
    return np.sqrt(s * (s - values[0]) * (s - values[1]) * (s - values[2]))

values = [float(input(f"Digite o valor {i+1}: ")) for i in range(3)]

print(f"Area do triangulo: {triangleAreaCalc(values):.2f}")
```

```python
#Exercicio 6

def biggerValue(a, b):
    return a if a > b else b

def diff(a, b):
    return a - b if a > b else b - a

a = int(input("Digite o valor 1: "))
b = int(input("Digite o valor 2: "))

print(f"Maior: {biggerValue(a, b):.2f} Diferença: {diff(a, b):.2f}")
```

```python
#Exercicio 7

def leapYearCalc(year):
    return "Bissexto" if year % 100 != 0 and year % 4 == 0 or year % 400 == 0 else "Nao Bissexto"

year = int(input("Digite o ano: "))

print(leapYearCalc(year))
```

```python
#Exercicio 8

def divisible(value):
    return True if value % 3 == 0 and value % 5 != 0 or value % 3 != 0 and value % 5 == 0 else False

value = int(input("Digite o valor: "))

print(divisible(value))
```

```python
#Exercicio 9

def retirementCalc(person):
    return "Pode aposentar" if person['gender'] == 'F' and person['age'] >= 60 or person['gender'] == 'M' and person['age'] >= 65 or person['workedYears'] >= 30 or person['age'] >= 60 and person['workedYears'] >= 25 else "Nao pode aposentar"

person = {
    'age': 0,
    'workedYears': 0,
    'gender': ''
}

person['age'] = int(input("Digite a idade: "))
person['gender'] = input("Digite o gênero (M ou F): ").upper()
person['workedYears'] = int(input("Digite os anos trabalhados: "))

print(retirementCalc(person))
```

```python
#Exercicio 10

def classificationCalc(person):
    if person['height'] <= 1.5:
        if person['weight'] <= 60:
            return "A"
        
        elif person['weight'] <= 90:
            return "D"
        
        else:
            return "G"
        
    elif person['height'] <= 1.8:
        if person['weight'] <= 60:
            return "B"
        
        elif person['weight'] <= 90:
            return "E"
        
        else:
            return "H"
        
    else:
        if person['weight'] <= 60:
            return "C"
        
        elif person['weight'] <= 90:
            return "F"
        
        else:
            return "I"

person = {
    'height': 0,
    'weight': 0
}

person['height'] = float(input("Digite a altura: "))
person['weight'] = float(input("Digite o peso: "))

print(classificationCalc(person))
```

```python
#Exercicio 11

import math

def geometric_mean(values):
    product = 1
    
    for value in values:
        product *= value
        
    return math.pow(product, 1 / len(values))

def weighted_mean(values, weights):
    weighted_sum = 0
    total_weight = 0
    
    for i in range(len(values)):
        weighted_sum += values[i] * weights[i]
        total_weight += weights[i]
        
    return weighted_sum / total_weight

def harmonic_mean(values):
    reciprocal_sum = 0
    
    for value in values:
        reciprocal_sum += 1 / value
        
    return len(values) / reciprocal_sum

def arithmetic_mean(values):
    return sum(values) / len(values)

def menu():
    option = ""
    
    while option != "E":
        print("\n(a) Geométrica")
        print("(b) Ponderada")
        print("(c) Harmônica")
        print("(d) Aritmética")
        print("(e) Sair")
        
        option = input("Digite a opção desejada: ").upper()
        
        if option in ["A", "B", "C", "D"]:
            values = [float(input(f"Digite o valor {i+1}: ")) for i in range(3)]
        
        match option:
            case "A":
                print(f"Média Geométrica: {geometric_mean(values)}")
                
            case "B":
                print(f"Média Ponderada: {weighted_mean(values, [1, 2, 3])}")
                
            case "C":
                print(f"Média Harmônica: {harmonic_mean(values)}")
                
            case "D":
                print(f"Média Aritmética: {arithmetic_mean(values)}")
                
            case "E":
                print("Saindo...")
                
            case _:
                print("Opção inválida! Tente novamente.")

menu()
```

```python
#Exercicio 12

def MDCCalc(x, y):
    while y != 0:
        x, y = y, x % y
        
    return x
    
x = int(input("Digite o valor 1: "))
y = int(input("Digite o valor 2: "))

print(MDCCalc(x, y))
```

```python
#Exercicio 13

def print_odds(N, M):
    if N % 2 == 0:
        N += 1
        
    else:
        N += 2
    
    for number in range(N, M, 2):
        print(number) 

N = int(input("Digite o valor de N: "))
M = int(input("Digite o valor de M: "))

if N >= M:
    print("N deve se menor que M")
        
else:
    print_odds(N, M)
```
